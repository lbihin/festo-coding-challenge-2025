## Objectif de la mission

L'agent bionique doit traverser les cartes dans l'ordre :

**Carte A → Carte B → Carte C**

La sortie (X) d'une carte est le début (S) de la suivante.

---

## Règles de déplacement

- Le déplacement se fait dans **4 directions** : N / E / S / O (pas de diagonales)
- Chaque déplacement traverse une tuile

---

## Types de tuiles

| Tuile | Signification | Remarques |
|-------|---------------|-----------|
| `#` | Mur | Impraticable |
| `.` | Route | Tuile standard |
| `~` | Terrain accidenté | Coût plus élevé |
| `M` | Montagne | Coût très élevé |
| `R` | Recharge | +10 énergie, +5 s pénalité temps |
| `K` | Clé | À ramasser, persiste pour cette carte |
| `>` | Convoyeur | Temps divisé par deux |
| `T` | Péage | Route + 2 énergie supplémentaire |
| `S` | Départ | Début du chemin |
| `X` | Sortie | Mène à la carte suivante |

### Règles spéciales

- La carte B contient le mécanisme clé-porte : **K → ouvre X dans la carte B**
- X dans la carte C est verrouillée jusqu'à la récupération de K en carte B

---

## Avantages des agents bioniques

| Agent | Bonus |
|-------|-------|
| **BionicOpter** | Multiplicateur de temps route (.) × 0,95 |
| **BionicSwift** | Coût en énergie Rough (~) réduit à 2 (au lieu de 3) |
| **BionicFlyingFox** | Coût Mountain (M) réduit à 3 (au lieu de 4) • Multiplicateur temps Mountain × 1,6 (au lieu de 1,8) |

---

## Coûts du terrain

### Coût en énergie par étape

| Tuile | Énergie |
|-------|---------|
| `.` | 1 |
| `~` | 3 |
| `M` | 4 |
| `T` | 3 (1 + 2 péage) |
| `R` | 1 (mais +10 énergie) |
| `>` | 1 |

### Coût en temps par pas

**Formule :**
```
Temps = base_step_time(Agent, Tile)
      × terrain_time_multiplier(Tile)
      × conveyor_factor
```

### Multiplicateurs de temps de terrain

| Tuile | Multiplicateur |
|-------|---|
| `.` | 1,0 |
| `~` | 1,4 |
| `M` | 1,8 |
| `T` | 1,0 |
| `R` | 1,0 |

### Facteur convoyeur

Sur les tuiles `>` uniquement : × 0,5

### Pénalité de recharge

Entrée en `R` : **+5 secondes** ajoutées au temps de pas normal

---

## Exemples de calcul

**BionicFlyingFox sur M :**
```
base_step_time = 1
Temps = 1 × 1,6 × 1 = 1,6 s
```

**BionicFlyingFox sur > (convoyeur) :**
```
base_step_time = 1
Temps = 1 × 1,6 × 0,5 = 0,8 s
```

**BionicOpter sur M :**
```
base_step_time = 0,95 (bonus temps)
Temps = 0,95 × 1,8 × 1 = 1,71 s
```

---

## Limites de temps

| Limite | Durée |
|--------|-------|
| Carte A | 20 s |
| Carte B | 20 s |
| Carte C | 20 s |
| **Total (toutes cartes)** | **36 s** |

---

## Format de sortie

Le chemin doit être une **seule chaîne NESW** combinée pour les 3 cartes.

**Exemple :**
```
SSEENW
```

**Signification :**
- `S, S` → se déplacer deux fois vers le sud
- `E, E` → se déplacer deux fois vers l'est
- `N` → se déplacer vers le nord
- `W` → se déplacer vers l'ouest

````python
DIRECTIONS = {
    "N" : (-1, 0),
    "S" : (1, 0),
    "E" : (0, 1),
    "W" : (0, -1)
}

def trouver_position(grille, ch):
    for l, ligne in enumerate(grille):
        c = ligne.find(ch)
        if c != -1:
            return (l, c)
    return
````

### Resolution du problème

````python
# Chargement du "monde"
import json
from pathlib import Path


data = json.loads(Path("res/world-model.json").read_text(encoding="utf-8"))

maps = data["maps"]
tiles = data["tiles"]
agents = data["agents"]
limits = data["limits"]

overall_limit_ms = int(limits["overall_seconds"] * 1000)

print(f"Monde chargé: {len(maps)} cartes, {len(tiles)} tuiles, {len(agents)} agents.")
````

````python
# Définition de la fonction "coût d´un pas" par agent
def cout_pas_ms(agent, map_key, tile_char):
    tile_info = tiles[tile_char]

    # Energie de base
    cout_energie = tile_info.get('energy', 0)

    # Modifie l´énergie de l´agent
    e_override = agent.get("perks", {}).get("energy_override", {})
    if tile_char in e_override:
        cout_energie = e_override[tile_char]

    # Multiplicateur du terrain
    mult_terrain = tile_info.get("time_mult", 1.0)
    # Override temps
    t_override = agent.get("perks", {}).get("time_mult_overrides", {})
    if tile_char in t_override:
        mult_terrain = t_override[tile_char]

    _b = agent["base_step_time"][map_key]

    conveyor_factor = tile_info.get("conveyor_time_factor", 1.0)
    temps_s = _b * mult_terrain * conveyor_factor

    # Pénalité recharge
    if "time_penalty" in tile_info:
        temps_s += float(tile_info["time_penalty"])
        
    # Retour en ms et énergie en int
    return int(cout_energie), int(round(temps_s * 1000))
    
````

````python
# Recherche du plus court chemin
import heapq


def dijkstra_one_map(map_idx, agent, require_key_from_prev_map=False):
    m = maps[map_idx]
    grid = m["grid"]
    map_key = chr(ord("A") + map_idx)  # "A", "B", "C"
    per_map_limit_ms = int(limits["per_map_seconds"][map_key] * 1000)

    start = trouver_position(grid, "S")
    goal  = trouver_position(grid, "X")
    has_key_in_this_map = False

    # Sur la Map B, X est bloqué tant que K n'est pas ramassée.
    # Sur la Map C, X est bloqué si on n'a pas eu la clé de la Map B (carry).
    is_map_b = (map_key == "B")
    is_map_c = (map_key == "C")
    map_requires_key_mech = bool(m["rules"].get("has_key", False))

    # énergie initiale de cette map
    energy0 = int(agent["battery"][map_key])

    # dist[(pos, have_key)] = (time_ms, energy, path)
    # On minimise time, et à temps égal on préfère plus d'énergie
    dist = {(start, False): (0, energy0, "")}
    heap = [(0, -energy0, start, False, "")]  # (time, -energy, pos, have_key, path)
    
    popped = 0

    while heap:
        t_ms, neg_e, pos, have_key, path = heapq.heappop(heap)
        popped += 1

        # si on a déjà mieux en stock, on skip
        if dist.get((pos, have_key)) != (t_ms, -neg_e, path):
            continue

        # condition d'arrivée
        if pos == goal:
            if map_requires_key_mech and not have_key:
                # On est sur X mais sans clé sur Map B : pas acceptable
                pass
            else:
                return (t_ms, -neg_e, path, have_key)

        r, c = pos
        for move, (dr, dc) in DIRECTIONS.items():
            nr, nc = r + dr, c + dc
            tile = grid[nr][nc]

            # mur ?
            if not tiles.get(tile, {}).get("passable", True) and tile != "X":
                continue
            if tile == "#":
                continue

            # logique de verrouillage
            if is_map_b and tile == "X" and not have_key:
                continue
            if is_map_c and tile == "X" and not require_key_from_prev_map:
                continue

            e_cost, dt_ms = cout_pas_ms(agent, map_key, tile)

            new_e = (-neg_e) - e_cost
            # recharge
            if "recharge" in tiles[tile]:
                new_e += int(tiles[tile]["recharge"])

            if new_e < 0:
                continue

            new_t = t_ms + dt_ms
            if new_t > per_map_limit_ms:
                continue

            new_have_key = have_key or tiles[tile].get("pickup_key", False)
            new_path = path + move
            state = ((nr, nc), new_have_key)

            best = dist.get(state)
            cand = (new_t, new_e, new_path)

            if best is None or (new_t < best[0]) or (new_t == best[0] and new_e > best[1]):
                dist[state] = cand
                heapq.heappush(heap, (new_t, -new_e, (nr, nc), new_have_key, new_path))

        # petit debug "humain" (tu peux commenter)
        if popped in (2000, 5000, 10000):
            print(f"[debug] {agent['name']} Map {map_key}: {popped} états explorés...")

    return # Rien trouvé
````

````python
def resoudre_pour_agent(agent):
    A = dijkstra_one_map(0, agent, require_key_from_prev_map=False)
    if A is None:
        print(f"Pas de solution pour {agent['name']} sur la Map A")
        return None
    
    B = dijkstra_one_map(1, agent, require_key_from_prev_map=False)
    if B is None:
        print(f"Pas de solution pour {agent['name']} sur la Map B")
        return None
    
    key_from_B = bool(B[3])
    C = dijkstra_one_map(2, agent, require_key_from_prev_map=key_from_B)
    if C is None:
        print(f"Pas de solution pour {agent['name']} sur la Map C")
        return None
    
    temps_total = A[0] + B[0] + C[0]
    chemin_total = A[2] + B[2] + C[2]
    return {
        "agent": agent["name"],
        "temps_ms": temps_total,
        "A": A, "B": B, "C": C,
        "chemin": chemin_total

    }
````

````python
import sys


solutions = []
for agent in agents:
    print(f"\n[+] Tentative avec {agent['name']}...")
    sol = resoudre_pour_agent(agent)
    if sol is None:
        print("    -> échec (pas de chemin ou contraintes trop strictes)")
        continue
    if sol["temps_ms"] <= overall_limit_ms:
        print(f"    -> OK ! total = {sol['temps_ms'] / 1000:.3f}s (<= {limits['overall_seconds']}s)")
    else:
        print(f"    -> chemin trouvé mais trop lent: {sol['temps_ms'] / 1000:.3f}s (> {limits['overall_seconds']}s)")
    solutions.append(sol)

# garde seulement celles qui respectent 36s
valid = [s for s in solutions if s["temps_ms"] <= overall_limit_ms]
if not valid:
    print("\n[-] Aucune solution ne respecte la limite globale. (Bizarre… vérifie les règles / JSON)")
    sys.exit(1)

best = min(valid, key=lambda s: s["temps_ms"])

print("\n==============================")
print("MEILLEURE SOLUTION VALIDE")
print("==============================")
print("Agent :", best["agent"])
print(f"Temps total : {best['temps_ms'] / 1000:.3f}s")
print("Chemin combiné (A+B+C) :")
print(best["chemin"])

# Si tu veux aussi les segments :
print("\n--- Segments ---")
print("Map A:", best["A"][2])
print("Map B:", best["B"][2])
print("Map C:", best["C"][2])
````
