# Le réveil bionique

Le monde vacille et se transforme autour de vous.

Vous vous trouvez à l'intérieur du Festo Technology Center. Au-dessus de vous, des AirJellies translucides flottent gracieusement dans les airs, élégantes, légères, presque vivantes. À proximité, des eMotionButterflies scintillantes virevoltent en formant des arcs délicats, leurs ailes traçant des trajectoires invisibles d'une grande précision. La salle entière ressemble à une cathédrale du mouvement et de l'ingénierie.

Pneumon flotte à côté de vous sous sa forme de méduse bionique, ses optiques brillant doucement.

> « L'ingénierie bio-inspirée », murmure-t-il. « Des machines qui bougent et s'adaptent comme des êtres vivants. Mais quelque chose ne va pas... »

Une à une, les AirJellies commencent à vaciller en plein vol. Les papillons entrent en collision, tourbillonnent et s'écrasent dans des éclats statiques. Une tempête de données corrompues se propage dans l'air : le Glitch est là, déformant leurs algorithmes de coordination.

Pneumon étend ses tentacules lumineuses et projette une grille de lumière sous vos pieds.

> « Leur logique de vol s'est fracturée en trois couches neuronales :
> - la stabilité
> - l'adaptation
> - la synchronisation
> 
> Nous devons guider chaque prototype à travers ces cartes corrompues pour rétablir l'équilibre. »

Trois silhouettes lumineuses se matérialisent à partir de la projection :

- **BionicOpter**
- **BionicSwift**
- **BionicFlyingFox**

Chacune doit naviguer dans une simulation à plusieurs niveaux :

| Carte | Objectif | Description |
|-------|----------|-------------|
| **A** | Stabilité | Conserver l'énergie, éviter les nœuds de surcharge, maintenir l'enveloppe de vol intacte. |
| **B** | Adaptation | Récupérer le nœud clé caché dans un motif changeant de courants pour déverrouiller la carte C. |
| **C** | Synchronisation | Les flux de données du convoyeur déferlent à une vitesse vertigineuse — suivez-les judicieusement ou risquez une décohérence totale. |

Chaque pas consomme de l'énergie, mais les nœuds de recharge ralentissent le temps. Équilibrez énergie et timing, sinon les prototypes sortiront de leur formation.

Pneumon fredonne nerveusement.

> « Vous devrez penser comme eux — glisser efficacement, vous adapter avec intention, vous déplacer comme un essaim. Ce n'est qu'ainsi que nous pourrons reconstruire ce que The Glitch a détruit. »

Le sol sous vos pieds pulse, formant le premier chemin lumineux.

> Un murmure d'ancien code résonne dans la salle : **« Mettez le mouvement avant l'effort dans la puissance »**.

---

## Votre tâche

- Guidez les prototypes à travers leurs grilles de vol
- Récupérez les données perdues
- Équilibrez l'énergie et le temps
- Restaurez l'harmonie au sein de l'essaim

Lorsque les trois seront à nouveau alignés, la salle brillera d'un rythme parfait — et un autre fragment d'histoire reprendra vie.

---

## Objectif de la mission

L'agent bionique doit traverser les cartes dans l'ordre :

**Carte A → Carte B → Carte C**

La sortie (X) d'une carte est le début (S) de la suivante.

---

## Règles de déplacement

- Le déplacement se fait dans **4 directions** : N / E / S / O (pas de diagonales)
- Chaque déplacement traverse une tuile

---

## Types de tuiles

| Tuile | Signification | Remarques |
|-------|---------------|-----------|
| `#` | Mur | Impraticable |
| `.` | Route | Tuile standard |
| `~` | Terrain accidenté | Coût plus élevé |
| `M` | Montagne | Coût très élevé |
| `R` | Recharge | +10 énergie, +5 s pénalité temps |
| `K` | Clé | À ramasser, persiste pour cette carte |
| `>` | Convoyeur | Temps divisé par deux |
| `T` | Péage | Route + 2 énergie supplémentaire |
| `S` | Départ | Début du chemin |
| `X` | Sortie | Mène à la carte suivante |

### Règles spéciales

- La carte B contient le mécanisme clé-porte : **K → ouvre X dans la carte B**
- X dans la carte C est verrouillée jusqu'à la récupération de K en carte B

---

## Avantages des agents bioniques

| Agent | Bonus |
|-------|-------|
| **BionicOpter** | Multiplicateur de temps route (.) × 0,95 |
| **BionicSwift** | Coût en énergie Rough (~) réduit à 2 (au lieu de 3) |
| **BionicFlyingFox** | Coût Mountain (M) réduit à 3 (au lieu de 4) • Multiplicateur temps Mountain × 1,6 (au lieu de 1,8) |

---

## Coûts du terrain

### Coût en énergie par étape

| Tuile | Énergie |
|-------|---------|
| `.` | 1 |
| `~` | 3 |
| `M` | 4 |
| `T` | 3 (1 + 2 péage) |
| `R` | 1 (mais +10 énergie) |
| `>` | 1 |

### Coût en temps par pas

**Formule :**
```
Temps = base_step_time(Agent, Tile)
      × terrain_time_multiplier(Tile)
      × conveyor_factor
```

### Multiplicateurs de temps de terrain

| Tuile | Multiplicateur |
|-------|---|
| `.` | 1,0 |
| `~` | 1,4 |
| `M` | 1,8 |
| `T` | 1,0 |
| `R` | 1,0 |

### Facteur convoyeur

Sur les tuiles `>` uniquement : × 0,5

### Pénalité de recharge

Entrée en `R` : **+5 secondes** ajoutées au temps de pas normal

---

## Exemples de calcul

**BionicFlyingFox sur M :**
```
base_step_time = 1
Temps = 1 × 1,6 × 1 = 1,6 s
```

**BionicFlyingFox sur > (convoyeur) :**
```
base_step_time = 1
Temps = 1 × 1,6 × 0,5 = 0,8 s
```

**BionicOpter sur M :**
```
base_step_time = 0,95 (bonus temps)
Temps = 0,95 × 1,8 × 1 = 1,71 s
```

## Format de sortie

Le chemin doit être une **seule chaîne NESW** combinée pour les 3 cartes.

**Exemple :**
```
SSEENW
```

**Signification :**
- `S, S` → se déplacer deux fois vers le sud
- `E, E` → se déplacer deux fois vers l'est
- `N` → se déplacer vers le nord
- `W` → se déplacer vers l'ouest
